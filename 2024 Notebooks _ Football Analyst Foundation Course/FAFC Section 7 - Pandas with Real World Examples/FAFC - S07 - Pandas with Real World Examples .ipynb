## More Advanced Groupbys

```python
# Convert 'position' to uppercase using map and lambda
pass_df['position_upper'] = pass_df['position'].map(lambda x: str(x).upper())

# Filter rows where 'duration' > 1 using lambda
filtered_df = pass_df[pass_df['duration'].map(lambda x: x > 1)]

# 2. Basics of `groupby` in Pandas

# Group by 'player' and get the mean of 'duration'
player_duration_mean = pass_df.groupby('player')['duration'].mean()

# Group by 'team' and 'player' and get the sum of 'pass_length'
team_player_pass_length = pass_df.groupby(['team', 'player'])['pass_length'].sum()

# 3. Using `lambda` with `groupby`

# Group by 'player' and calculate the range (max - min) of 'duration'
player_duration_range = pass_df.groupby('player')['duration'].agg(lambda x: x.max() - x.min())

# Filter groups (players) where the average 'duration' is more than 1.5
filtered_groups = pass_df.groupby('player').filter(lambda x: x['duration'].mean() > 1.5)

# Subtract each player's 'duration' by their mean 'duration'
centered_duration = pass_df.groupby('player')['duration'].transform(lambda x: x - x.mean())

print(centered_duration)
```
