# OOP - Intro to Object Oriented Programming

- Object-Oriented Programming (OOP) is a programming paradigm based on the concept of "objects," which can contain data in the form of fields (often known as attributes or properties) and code in the form of procedures (often known as methods). 

- OOP models real-world or abstract entities as objects with states and behaviors.

- In OOP, each object is an instance of a class. A class can be thought of as a blueprint or template for creating objects. It defines the properties and behaviors that the objects created from the class will have.


Why OOP?

OOP allows for more manageable, flexible, and organized code, especially when dealing with complex systems. It enables developers to create objects that are self-contained within their own class, making it easier to debug, maintain, and expand upon.

## Key Concepts of OOP

- **Class:** A blueprint for creating objects (a particular data structure), providing initial values for state (member variables or attributes), and implementations of behavior (member functions or methods).

- **Object:** A specific instance of a class. It has a state and behavior defined by its class.

- **Method:** A function defined in a class that describes the behaviors of the objects created from the class.

- **Inheritance:** The mechanism by which one class can inherit the attributes and methods of another class.

- **Encapsulation:** The bundling of data and the methods that operate on that data within one unit. This restricts the access to some of the object's components, which is a means of preventing accidental interference and misuse of the methods and data.

- **Polymorphism:** The ability for different classes to be treated as instances of the same class through inheritance. It allows for methods to be created which can be used by objects from different classes.

The compare_goals method is designed to evaluate a player's performance relative to their peers. It takes a list of goals scored by other players as input, and its primary objective is to determine the player's ranking in the race for the title of the best striker.


Unlike methods such as update_stats, which internally modify a player's statistics, compare_goals is externally focused. It assesses how a player's performance stacks up against others in a specific aspect – goal-scoring ability.

**Input:** While methods like update_stats require data like additional games played and goals scored to update the player's attributes, compare_goals relies on external data in the form of a list of goals scored by other players.

**Outcome:** The result of calling compare_goals is not an update to the player's attributes; instead, it returns a ranking. This ranking represents where the player stands in terms of goal-scoring compared to their peers.

```python
class Player:
    def __init__(self, last_name, first_name, age, games, goals):
        self.last_name = last_name
        self.first_name = first_name
        self.age = age
        self.games = games
        self.goals = goals
        self.goals_per_game = self.goals / self.games if self.games > 0 else 0

    def update_stats(self, additional_games, additional_goals):
        self.games += additional_games
        self.goals += additional_goals
        self.goals_per_game = self.goals / self.games if self.games > 0 else 0

    def compare_goals(self, other_players_goals):
        # Create a list that includes the current player's goals and other players' goals
        all_goals = other_players_goals + [self.goals]

        # Sort the list in descending order so more goals have a higher rank
        all_goals.sort(reverse=True)

        # Find the rank of the current player (index in the sorted list plus one)
        player_rank = all_goals.index(self.goals) + 1

        return player_rank
```

## Class inheritance

Class inheritance is a fundamental concept in object-oriented programming where a new class (subclass) is derived from an existing class (superclass), inheriting its attributes and methods while also having the ability to introduce its own unique features or modify the inherited ones. 

This allows for code reuse, hierarchical organization, and a more efficient way to represent real-world relationships in programming.

### Code Reusability:

- Inheritance allows us to write common attributes and methods only once in a base class and then reuse them in derived classes, reducing code duplication and the potential for errors.

### Extensibility:

- It's easy to add new features or modify existing behavior in one place (the base class), and those changes can automatically apply to all derived classes.

### Representing Real-World Relationships:

- Inheritance enables us to model real-world relationships in code. For example, a Goalkeeper is a type of Player, so it makes sense for a Goalkeeper class to inherit from a Player class.

Lets create a goalkeepr class

Goalkeeper is a type of a player, so it takes all arguments of the parent class Player, but adds an argument that is specific to Goalkeeper - clean_sheets

```python
class Goalkeeper(Player):
    def __init__(self, last_name, first_name, age, games, goals, clean_sheets):
        super().__init__(last_name, first_name, age, games, goals)
        self.clean_sheets = clean_sheets
```
