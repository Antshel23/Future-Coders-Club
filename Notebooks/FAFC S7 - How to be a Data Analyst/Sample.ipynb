```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#ARTETA DATA
```

```python
data = {
    'Season': ['19/20', '20/21', '21/22', '22/23', '23/24', '24/25'],
    'xG': [1.31, 1.37, 1.70, 1.87, 2.14, 2.0],
    'Opponent xG': [1.53, 1.20, 1.24, 1.13, 0.90, 0.97],
    'Possession (%)': [53.6, 53.0, 53.0, 59.1, 57.5, 54.8]
}
```

```python
#CONVERT DATA OBJECT TO DATAFRAME
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
#BASIC CALCULATIONS
```

```python
df['xG Difference'] = df['xG'] - df['Opponent xG']
```

```python
print(df)
```

```python
average_opponent_xG = df['Opponent xG'].mean()
```

```python
print(average_opponent_xG)
```

```python
#LOGICAL STATEMENTS E.G. IF ELIF ELSE, LOOPS E.G. FOR, WHILE
```

```python
for index, row in df.iterrows():
    if row['xG Difference'] < 0:
        print(f"Season: {row['Season']} - negative")
    else:
        print(f"Season: {row['Season']} - positive")
```

```python
#BASIC VIZ
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Season'], df['xG Difference'], marker='o', linestyle='--', color='r', label='xG Difference')
plt.axhline(0, color='r', linestyle='--', linewidth=1)
plt.xlabel('Season')
plt.ylabel('xG Difference')
plt.title('xG Difference Over Seasons')
plt.legend()
plt.show()
```
